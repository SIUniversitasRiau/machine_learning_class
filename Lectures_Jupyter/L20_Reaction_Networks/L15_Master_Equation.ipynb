Master equation: The probability that if we start it off with a some labelling at time $t_1$, our stochastic Petri net will wind up with some other labelling at time $t_2$

We sum over Feynman diagrams that start and end with the given labellings.

For each Feynman diagram, we integrate (the product of the rate constants) over all possible times at which the transitions occur.

$$\frac{d}{dt}\phi_{l^\prime}=\sum_l H_{l^\prime l}\phi_l(t)$$

A labelling of this stochastic Petri net is a k-tuple of natural numbers $(l_1,\cdots,l_k)$ saying how many things there are of each species. 

Considering a vector space with labellings $l$ as basis, then $\phi_l(t)$ will be the $l$ component of a vector and $H_{l^\prime l}$ will be the matrix entries of an operator.

$$\frac{d}{dt}\phi(t)=H\phi(t)$$

The master equation describes probabilities while in quantum theory, the Schrondinger equation deals with amplitudes.

Exercise. Suppose we have $k$ species, $1$ transition whose rate constant is $r$. Suppose the $i$th species appears $m_i$ times as input and $n_i$ times as output.

The probability that the transition occurs in a short moment $\delta t$ is approximately:

the rate constant $r$ times 

the time $\delta t$ times

the number of ways the transition can occur $l_1^{m_1}\cdots l_k^{m_k}=l^m$

Then the rate at which the transition occurs is: $r\times l^m$. When it occurs, it takes $m_i$ things of the $i^{th}$ species and produces $n_i$ things of that species. Then $$\frac{d}{dt}\phi_{l-m+n}= rl^{m}, H_{l-m+n\,\,\, l}=rl^{m}.$$
