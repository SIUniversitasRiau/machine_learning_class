<font size='6' style="color:#00A6D6">  <b>Parallel Planes</b></font>

Given two planes, in how many points can they intersect?

In $R^3$ we have three options, they can not intersect, intersect in a line, or be the same plane.

But in $R^4$ things are more complicated, as they can intersect in a point.  More over the examples are easy to construct!

$A=\{(a,b,0,0)|\,a,b\in R\}$ and $B=\{(0,0,c,d)|\,c,d\in R\}$ they only share the point $(0,0,0,0)$.

<font size='6' style="color:#00A6D6">  <b>Density</b></font> 

Most of the volume of the hypercube is near the borders. Consider a hypercube of side 1. If you cut a slice of the cube near a face, of volume $1x1x\epsilon$, then the higher the dimension, the more faces the hypercube have, and so, even with the same volume, those slices amount for a higher percentage of the total volume. 

It can be shown that most of the surface area of the hyper sphere is around the equator.

    If a constant number of examples is distributed uniformly in a high dimensional hypercube,
    beyond some dimensionality most examples are closer to a face of the hypercube than to their
    nearest neighbor.

We consider a ball of radious 1 inside a cube of side 2. The ball touches the sides. 

What fraction of the cube is filled by the ball?

    In one dimension, the ball and the cube coincide, so the answer is 1.

    In two dimensions, a 2-ball inside a 2-cube is a disk inside a square, so the ratio is about .79.

    In three dimensions the ratio is approximately .52.

In general the hyper cube has volume $2^d$, while the volume of the hyper sphere of radious 1 is given by:
$\frac{\pi^{d/2}}{\Gamma(d/2+1)}$, this gamma function is an extension of the factorial:
$\Gamma(n+1)=n!$ for non integrers, so we can ask things like what is $\Gamma(1/2)? \sqrt{\pi}.$

<font size='6' style="color:#00A6D6">  <b>KNN</b></font>  

kNN in low dimensions have to select among the closest points. In higher dimensions there may be too many points nearby and so the choice seems random.  Actually, there is a more serious problem for KNN.


Let's work with the unit hypercube. Fix a random corner. We find that there are n corners with in distance 1. ${n\choose 2}$ with in distance $\sqrt{2}$, ${n\choose 2}$ with in distance $\sqrt{3},\cdots, {n \choose n/2}$ corners within distance $\sqrt{n/2}\cdots$ until one last point with distance $\sqrt{n}.$


We know that ${n \choose n/2}$ is the biggest binomial coefficient, so it contains most corners, more over the values around $\sqrt{n}$ differ by little to it, so the distribution of points by distance is  skew and centered near that value.

```python
from IPython.display import Image
Image("dimensionality/dim.png")
```

It can be shown that if the data is independent and identically distributed, the ratio between the nearest and farthest points approaches most of the time, i.e. the points essentially become uniformly distant from each other. This phenomenon can be observed for wide variety of distance metrics, but it is more pronounced for the Euclidean metric. 

On the other hand, among all possible $L^k$ metrics, the $L^1$ and $L^2$ are better.

Fortunately, there is an effect that partly counteracts the
curse, which might be called the“blessing of non-uniformity.”
In most applications examples are not spread uniformly throughout
the instance space, but are concentrated on or near
