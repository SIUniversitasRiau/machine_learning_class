```python
import numpy as np
from tqdm import tqdm_notebook
import matplotlib.pyplot as plt
import cmath
from copy import copy
```

In this exercise we are going to place 100 points in the map:

```python
h,w, maxit=500,500,50
"""Returns an image of the Mandelbrot fractal of size (h,w)."""
X = np.ones((500,500))
c = (-0.8+0.156*1j)*X
y,x = np.ogrid[ -1.5:1.5:h*1j, -1:1:w*1j ]
z= x+y*1j


Julia = maxit + np.zeros(z.shape, dtype=int)

for i in tqdm_notebook(range(maxit)):
    z = z**2+c 
    diverge = (z)*np.conj(z) > 2**2            # who is diverging
    div_now = diverge & (Julia==maxit)  # who is diverging now
    Julia[div_now] = i                  # note when
    z[diverge] = 2                        # avoid diverging too much


fig = plt.figure()
ax = fig.add_axes([0., 0., 1., 1., ])
# Hide grid lines
ax.grid(False)
ax.axis('off')
# Hide axes ticks
plt.imshow(Julia)


plt.show()
```

And they will cluster around the more colorful points using SOM.

```python
#points = []
class node():
    def __init__(self,x,y,v):
        self.c = x+y*1j
        self.v = v
nodes = np.zeros((10,10),dtype=object)

for x_values in range(0,10):
    for y_values in range(0,10):
        nodes[x_values][y_values]=node(50*x_values,50*y_values,Julia[50*x_values,50*y_values])
        #Here the first copy is the colors, an the second the coordinates
```

```python
fig = plt.figure()

ax = fig.add_axes([0., 0., 1., 1., ])
# Hide grid lines
ax.grid(False)
ax.axis('off')
# Hide axes ticks

plt.imshow(Julia)

copyNodes = np.ravel(nodes)
for node in tqdm_notebook(copyNodes):
    plt.scatter(node.c.real, node.c.imag, s=5, c='red', marker='o')

plt.show()
```

```python
def distance(z,w):
    return (z[0]-w[0])**2+(z[1]-w[1])**2
def findNbh(indexX,indexY,nodes):
    copyNodes=np.ravel(nodes)
    OrderedCopyNodes = sorted(copyNodes, key=lambda item: abs(item.c- nodes[indexX][indexY].c))
    return OrderedCopyNodes[:5]
def findBMU(indexX,indexY,nodes):
    copyNodes=np.ravel(nodes)
    OrderedCopyNodes = sorted(copyNodes, key=lambda item: abs(item.v- nodes[indexX][indexY].v))
    return OrderedCopyNodes[:5][1]
```

```python
firstNode = findNbh(5,5,nodes)
print([x.c for x in firstNode])
```

```python
secondNode = findBMU(5,5,nodes)
print(secondNode.c.real)
numberOfIterations = 10
```

```python
for iteration in tqdm_notebook(range(numberOfIterations)):
    for x_values in range(0,10):
        for y_values in range(0,10):
            BMUc = findBMU(x_values,y_values,nodes).c
            for nearNodes in findNbh(x_values,y_values,nodes):
                newCoord  = nearNodes.c + .1*BMUc
                newX, newY=int(newCoord.real)%500, int(newCoord.imag)%500
                nearNodes.c = newX + newY*1j            
                #nearNodes.c = nearNodes.c + .1*BMUc
                nearNodes.v = Julia[newX, newY]
```

```python
Julia2 = copy(Julia)
fig = plt.figure()

ax = fig.add_axes([0., 0., 1., 1., ])
# Hide grid lines
ax.grid(False)
ax.axis('off')
# Hide axes ticks

plt.imshow(Julia2)

copyNodes = np.ravel(nodes)
for node in copyNodes:
    plt.scatter(node.c.real, node.c.imag, s=5, c='red', marker='o')


plt.show()
```
